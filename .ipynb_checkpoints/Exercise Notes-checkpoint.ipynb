```python
# first, look at main.py, it is quite simple:
#from agent import Agent #import agent
#from monitor import interact #monitor is where things will run
import gym #this is the environment
import numpy as np
import random
from collections import deque #deque is like a list, optimized for 
import sys
import math
from collections import defaultdict
```

```python
#contents of monitor.py, don't change
def interact(env, agent, num_episodes=20000, window=100):
    """ Monitor agent's performance.
    
    Params
    ======
    - env: instance of OpenAI Gym's Taxi-v1 environment
    - agent: instance of class Agent (see Agent.py for details)
    - num_episodes: number of episodes of agent-environment interaction
    - window: number of episodes to consider when calculating average rewards

    Returns
    =======
    - avg_rewards: deque containing average rewards
    - best_avg_reward: largest value in the avg_rewards deque
    """
    # initialize average rewards. looks like an empty list
    avg_rewards = deque(maxlen=num_episodes)
    # initialize best average reward
    best_avg_reward = -math.inf # i guess the worst reward is negative infinity
    # initialize monitor for most recent rewards
    samp_rewards = deque(maxlen=window)
    # for each episode
    # each episode so far starts like this
    for i_episode in range(1, num_episodes+1): #standard
        # begin the episode
        state = env.reset() #standard
        # initialize the sampled reward
        samp_reward = 0 #just set the first value
        while True:
            # agent selects an action
            action = agent.select_action(state)
            # agent performs the selected action
            next_state, reward, done, _ = env.step(action)
            # agent performs internal updates based on sampled experience
            agent.step(state, action, reward, next_state, done)
            # update the sampled reward
            samp_reward += reward
            # update the state (s <- s') to next time step
            state = next_state
            if done:
                # save final sampled reward
                samp_rewards.append(samp_reward)
                break
        if (i_episode >= 100):
            # get average reward from last 100 episodes
            avg_reward = np.mean(samp_rewards)
            # append to deque
            avg_rewards.append(avg_reward)
            # update best average reward
            if avg_reward > best_avg_reward:
                best_avg_reward = avg_reward
        # monitor progress
        print("\rEpisode {}/{} || Best average reward {}".format(i_episode, num_episodes, best_avg_reward), end="")
        sys.stdout.flush()
        # check if task is solved (according to OpenAI Gym)
        if best_avg_reward >= 9.7:
            print('\nEnvironment solved in {} episodes.'.format(i_episode), end="")
            break
        if i_episode == num_episodes: print('\n')
    return avg_rewards, best_avg_reward
```

```python
#contents of agent.py, need to change

class Agent:

    def __init__(self, nA=6, algo = 'expsarsa'):
        """ Initialize agent.

        Params
        ======
        - nA: number of actions available to the agent
        """
        self.nA = nA
        self.Q = defaultdict(lambda: np.zeros(self.nA))
        self.gamma = 0.9 #was 1.0
        self.eps_start = 1.0
        self.epsilon = self.eps_start
        self.eps_decay = 0.99999
        self.eps_min = 0.00005 #was 0.05
        #self.eps = 0.005
        self.alpha = 0.9
        self.algo = algo
        
    def select_action(self, state): #using epsilon greedy
        """ Given the state, select an action.

        Params
        ======
        - state: the current state of the environment

        Returns
        =======
        - action: an integer, compatible with the task's action space
        """
        #return np.random.choice(self.nA) #just a random selection
        if random.random() > self.epsilon: # select greedy action with probability epsilon
            return np.argmax(self.Q[state])
        else:                     # otherwise, select an action randomly
            return random.choice(np.arange(self.nA))

    def step(self, state, action, reward, next_state, done):
        """ Update the agent's knowledge, using the most recently sampled tuple.

        Params
        ======
        - state: the previous state of the environment
        - action: the agent's previous choice of action
        - reward: last reward received
        - next_state: the current state of the environment
        - done: whether the episode is complete (True or False)
        """
        #i think this only updates Q
        #self.Q[state][action] += 1
        # prepare for discounting
        current = self.Q[state][action]
        

        if self.algo == 'sarsamax':
        #here we are using update_Q_sarsamax        
            Qsa_next = np.max(self.Q[next_state]) if next_state is not None else 0  # value of next state 
            target = reward + (self.gamma * Qsa_next)               # construct TD target
            new_value = current + (self.alpha * (target - current)) # get updated value 
            
        elif self.algo == 'sarsa':
        #here we are using update_Q_sarsa
        #including epsilon greedy algo
            if random.random() > self.epsilon: # select greedy action with probability epsilon
                next_action = np.argmax(self.Q[state])
            else:                     # otherwise, select an action randomly
                next_action = random.choice(np.arange(self.nA))

            Qsa_next = self.Q[next_state][next_action] if next_state is not None else 0    
            target = reward + (self.gamma * Qsa_next)               # construct TD target
            new_value = current + (self.alpha * (target - current))

        else:
        #here we are using update_Q_expsarsa
            policy_s = np.ones(self.nA) * self.epsilon / self.nA
            policy_s[np.argmax(self.Q[next_state])] = 1 - self.epsilon + (self.epsilon / self.nA)
            Qsa_next = np.dot(self.Q[next_state], policy_s) 
            new_value = current + (self.alpha * (reward + (self.gamma * Qsa_next) - current)) #update_Q_expsarsa
        
        self.Q[state][action] = new_value
        
        #let's also do epsilon decay here
        if self.epsilon >= self.eps_min:
            self.epsilon = self.epsilon * self.eps_decay
        else:
            self.epsilon = self.eps_min
```

```python
#contents of main.py. should be able to run these and be done

env = gym.make('Taxi-v2') #initializing the environment
agent = Agent(algo = 'expsarsa') #initializing the agent
avg_rewards, best_avg_reward = interact(env, agent, num_episodes = 200000) #using the interact function in monitor.py, we will begin to generate average and best average rewards

#OpenAI Gym defines "solving" this task as getting average return of 9.7 over 100 consecutive trials
#record keeping
#Episode 20000/20000 || Best average reward 7.359
#Episode 200000/200000 || Best average reward 8.04
#Episode 200000/200000 || Best average reward 9.22
#Episode 200000/200000 || Best average reward 9.38
#Episode 200000/200000 || Best average reward 9.45
#Episode 200000/200000 || Best average reward 9.35

#Episode 200000/200000 || Best average reward 9.39

#Episode 200000/200000 || Best average reward 9.49

```

```python
#some notes
#lets look at the three algos
#SARSA0, or SARSA
def update_Q_sarsa(alpha, gamma, Q, state, action, reward, next_state=None, next_action=None):
    """Returns updated Q-value for the most recent experience."""
    current = Q[state][action]  # estimate in Q-table (for current state, action pair)
    # get value of state, action pair at next time step
    Qsa_next = Q[next_state][next_action] if next_state is not None else 0    
    target = reward + (gamma * Qsa_next)               # construct TD target
    new_value = current + (alpha * (target - current)) # get updated value
    return new_value

#QLearning, or SARSAMAX
def update_Q_sarsamax(alpha, gamma, Q, state, action, reward, next_state=None):
    """Returns updated Q-value for the most recent experience."""
    current = Q[state][action]  # estimate in Q-table (for current state, action pair)
    Qsa_next = np.max(Q[next_state]) if next_state is not None else 0  # value of next state 
    target = reward + (gamma * Qsa_next)               # construct TD target
    new_value = current + (alpha * (target - current)) # get updated value 
    return new_value

#Expected SARSA
def update_Q_expsarsa(alpha, gamma, nA, eps, Q, state, action, reward, next_state=None):
    """Returns updated Q-value for the most recent experience."""
    current = Q[state][action]         # estimate in Q-table (for current state, action pair)
    policy_s = np.ones(nA) * eps / nA  # current policy (for next state S')
    policy_s[np.argmax(Q[next_state])] = 1 - eps + (eps / nA) # greedy action
    Qsa_next = np.dot(Q[next_state], policy_s)         # get value of state at next time step
    target = reward + (gamma * Qsa_next)               # construct target
    new_value = current + (alpha * (target - current)) # get updated value 
    return new_value


'''
all three started by initiating the current value to Q[state][action]
next step:
SARSA0: look at the next Q value = Q[next_state][next_action], where the next_action and next_states are specified
SARSAMAX: Q value = np.max(Q[next_state]), where policy is geared towards the next_state's best action
EXPECTEDSARSA: Q value = np.dot(Q[next_sate], policy), where policy is geared towards the next_state's best action, and with probability involved

then the calculation for the new value are similar, in that
target = reward + gamma * Q next
new_value = current + alpha * (target - current)
'''
```
